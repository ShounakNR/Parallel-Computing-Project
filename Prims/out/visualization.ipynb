```python
import pandas as pd
```

```python
df = pd.read_csv("4threads.txt", header = None)
```

```python
df.columns = ['size_of_warp','time_seq','time_par','threads']
```

```python
df.to_csv("4threads.csv", index=None)
```

```python
df2 = pd.read_csv("2threads.txt", header = None)
```

```python
df2.columns = ['size_of_warp','time_seq','time_par','threads']
```

```python
df2.to_csv("2threads.csv", index = None)
```

```python
import matplotlib.pyplot as plt
```

```python
size = df2['size_of_warp']
tseq = df2['time_seq']
tpar = df2['time_par']
```

```python
plt.plot(size, tseq, color='red')
plt.plot(size, tpar, color='blue')
plt.ylabel('Time')
plt.xlabel('Red: Seq, Blue: Par')
plt.title('2 Threads')
plt.show()
```

```python
size4 = df['size_of_warp']
tseq4 = df['time_seq']
tpar4 = df['time_par']
```

```python
plt.plot(size4, tseq4, color='red')
plt.plot(size4, tpar4, color='blue')
plt.ylabel('Time')
plt.xlabel('Red: Seq, Blue: Par')
plt.title('4 Threads')
plt.show()
```

```python
speed_up = []
for _ in range(len(tseq)):
    speed_up.append(tseq[_]/tpar[_])
```

```python
plt.plot(size, speed_up, color='red')
# plt.plot(size4, tpar4, color='blue')
plt.ylabel('Time')
plt.xlabel('Red:Speed Up')
plt.title('2 Threads')
plt.show()
```
